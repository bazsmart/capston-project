```python
import pandas as pd
```

```python
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])
# Calculate ride length in minutes
df['ride_length'] = (df['ended_at'] - df['started_at']).dt.total_seconds()/ 60
# Add day of week (1=Sunday, 7=saturday)
df['day_of_week'] = df['started_at'].dt.dayofweek.map(
    {0:2, 1:3, 2:4, 3:5, 4:6, 5:7, 6:1}
)
print(df[['started_at','ended_at','ride_length','day_of_week']].head())
```

```python
# Remove negative or zero ride lengths
df = df[df['ride_length']>0]
# Check whats left
print(df.shape)
```

```python
print(df['member_casual'].value_counts())
```

```python
print(df.groupby('member_casual')['ride_length'].mean().round(2))
```

```python
print(df.groupby(['member_casual','day_of_week'])['ride_id'].count())
```

```python
# Check for nulls in key columns
print(df[['ride_length','member_casual','day_of_week']].isnull().sum())
```

```python
# Check for rides over 24 hours (likely docking errors)
print(df[df['ride_length'] > 1440].shape[0])
```

```python
df = df[df['ride_length'] <= 1440]
print(df.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
df['member_casual'].value_counts().plot(kind='bar', color=['#1a7abf','#f4a11d'], edgecolor='white')
plt.title('Total Rides: Members vs Casual Riders')
plt.xlabel('Rider Type')
plt.ylabel('Number of Rides')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('chart1_total_rides.png')
plt.show()
```

```python
plt.figure(figsize=(8,5))
df.groupby('member_casual')['ride_length'].mean().plot(kind='bar', color=['#1a7abf','#f4a11d'], edgecolor='white')
plt.title('Average Ride Length: Members vs Casual Riders')
plt.xlabel('Rider Type')
plt.ylabel('Average Ride Length (minutes)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('chart2_avg_ride_length.png')
plt.show()
```

```python
day_labels = {1:'Sun', 2:'Mon', 3:'Tue', 4:'Wed', 5:'Thu', 6:'Fri', 7:'Sat'}

rides_by_day = df.groupby(['member_casual','day_of_week'])['ride_id'].count().reset_index()
rides_by_day['day_name'] = rides_by_day['day_of_week'].map(day_labels)

plt.figure(figsize=(10,5))
for rider, color in zip(['member','casual'], ['#1a7abf','#f4a11d']):
    data = rides_by_day[rides_by_day['member_casual']==rider]
    plt.plot(data['day_name'], data['ride_id'], marker='o', label=rider, color=color, linewidth=2.5)

plt.title('Number of Rides by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Rides')
plt.legend(title='Rider Type')
plt.tight_layout()
plt.savefig('chart3_rides_by_day.png')
plt.show()
```

```python
avg_by_day = df.groupby(['member_casual','day_of_week'])['ride_length'].mean().reset_index()
avg_by_day['day_name'] = avg_by_day['day_of_week'].map(day_labels)

plt.figure(figsize=(10,5))
for rider, color in zip(['member','casual'], ['#1a7abf','#f4a11d']):
    data = avg_by_day[avg_by_day['member_casual']==rider]
    plt.plot(data['day_name'], data['ride_length'], marker='o', label=rider, color=color, linewidth=2.5)

plt.title('Average Ride Length by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Average Ride Length (minutes)')
plt.legend(title='Rider Type')
plt.tight_layout()
plt.savefig('chart4_avg_length_by_day.png')
plt.show()
```

```python
summary = df.groupby(['member_casual','day_of_week']).agg(
    total_rides=('ride_id','count'),
    avg_ride_length=('ride_length','mean')
).round(2).reset_index()

summary.to_csv('cyclistic_summary.csv', index=False)
print('Done')
```
